```python
##Funcion que genera una lista de n numeros aletorios enteros dentro del intervalo (a,b)
```

```python
def randint(n,a,b):
    x=[]
    for i in range(n):
        x.append(np.random.randint(a,b))
    return x
```

```python
randint(25,1,50)
```

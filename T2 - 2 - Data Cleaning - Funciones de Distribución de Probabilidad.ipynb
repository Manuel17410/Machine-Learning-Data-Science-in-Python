## Distribucion Uniforme

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a=1
b=100
n=10000

data=np.random.uniform(a,b,n)
```

```python
plt.hist(data)
```

## Distribucion Normal

```python
data=np.random.randn(10000)
```

```python
x= range(1,10001)
```

```python
plt.plot(x,data)
```

```python
plt.hist(data)
```

```python
plt.plot(x,sorted(data))
```

```python
mu=5.5
std=2.5
Z_10000 = mu + std*np.random.randn(10000)
plt.hist(Z_10000)
```

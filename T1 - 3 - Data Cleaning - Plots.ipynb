# Plot y visualizacion de los datos

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\Asus\Desktop\Datasets\datasets\customer-churn-model\Customer Churn Model.txt")
```

```python
data.head()
```

## Scatter Plot

```python
data.plot(kind="scatter", x="Day Mins", y= "Day Charge")
```

```python
data.plot(kind="scatter", x="Night Mins", y= "Night Charge")
```

```python
figure, axs = plt.subplots(2,2,sharey=True, sharex=True)
data.plot(kind="scatter", x="Day Mins", y= "Day Charge", ax=axs[0][0])
data.plot(kind="scatter", x="Night Mins", y= "Night Charge", ax=axs[1][0])
```

## Histograma de Frecuencias

```python
plt.hist(data["Day Calls"])
```

```python
plt.hist(data["Day Calls"], bins=20)
plt.xlabel("Numero de llamadas al dia")
plt.ylabel("Frecuencia")
plt.title("Histograma de numero de llamadas al dia")
```

```python
plt.hist(data["Day Calls"], bins=[30,60,90,120,150,180])
plt.xlabel("Numero de llamadas al dia")
plt.ylabel("Frecuencia")
plt.title("Histograma de numero de llamadas al dia")
```

## Boxplot

```python
plt.boxplot(data["Day Calls"])
plt.title("Box Plot de llamadas diarias")
plt.ylabel("Day Calls")
```

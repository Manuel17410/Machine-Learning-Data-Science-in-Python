# Distancias

```python
from scipy.spatial import distance_matrix
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\Asus\Desktop\Datasets\datasets\movies\movies.csv", sep=";")
```

```python
data.head()
```

```python
movies = data.columns.values.tolist()[1:]
movies
```

```python
dd1 = distance_matrix(data[movies],data[movies],p=1)
dd2 = distance_matrix(data[movies],data[movies],p=2)
dd10 = distance_matrix(data[movies],data[movies],p=10)
```

```python
dd1
```

```python
def dm_to_df(dd,col_name):
    import pandas as pd
    return pd.DataFrame(dd,index=col_name,columns=col_name)
```

```python
dm_to_df(dd1,data["user_id"])  #manhattan
```

```python
dm_to_df(dd2,data["user_id"])  #euclidea
```

```python
dm_to_df(dd10,data["user_id"])  #minkowsky
```

## Clustering Jerarquico

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
movies
```

```python
data[movies]
```

```python
Z = linkage(data[movies],"ward")
Z
```

```python
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerarquico para el clusterint")
plt.xlabel("Distancias")
plt.ylabel("ID de peliculas")
dendrogram(Z,leaf_rotation=90.,leaf_font_size=10.)
plt.show()
```

```python
Z = linkage(data[movies],"average")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerarquico para el clusterint")
plt.xlabel("Distancias")
plt.ylabel("ID de peliculas")
dendrogram(Z,leaf_rotation=90.,leaf_font_size=10.)
plt.show()
```

```python
Z = linkage(data[movies],"complete")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerarquico para el clusterint")
plt.xlabel("Distancias")
plt.ylabel("ID de peliculas")
dendrogram(Z,leaf_rotation=90.,leaf_font_size=10.)
plt.show()
```

```python
Z = linkage(data[movies],"single")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerarquico para el clusterint")
plt.xlabel("Distancias")
plt.ylabel("ID de peliculas")
dendrogram(Z,leaf_rotation=90.,leaf_font_size=10.)
plt.show()
```
